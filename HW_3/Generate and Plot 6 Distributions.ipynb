```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as pl
import json
s = json.load(open('fbb_matplotlibrc.json'))
pl.rcParams.update(s)
```

```python
from __future__ import print_function
__author__= 'yiningfan'
import os

from scipy.optimize import curve_fit, minimize
```

```python
# Generating 100 different normal distribution with same sample mean and different sample size
# np.random.normal(mean, sigma, size)
md={}
means = {}
mymean = 100
mysize=2000/(np.array(range(1,100)+[10]))
np.random.seed(2015)
md['normal']={}
mu=mymean
sigma=1
for n in mysize:
    md['normal'][n] = np.random.normal(mu,sigma,n)
means['normal'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['normal'].iterkeys():
    means['normal'][nn] = md['normal'][nn].mean()
    axchisq_mu_n.plot(nn, means['normal'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')

# Generating 100 exponential distributions with same sample mean and different sample size
# np.random.exponential(bt, size)
np.random.seed(2015)
md['exp']={} 
bt = mymean
for n in mysize:
    md['exp'][n] = np.random.exponential(bt, size=n)
    
means['exp'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['exp'].iterkeys():
    means['exp'][nn] = md['exp'][nn].mean()
    axchisq_mu_n.plot(nn, means['exp'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')
```

```python
# Generating 100 differnt Poisson Distribution with same sample mean and different sample size
# np.random.poisoon(lamda, size)
# lamda = mean
np.random.seed(2015)
md['poisson']={}
ld=mymean
for n in mysize:
    md['poisson'][n] = np.random.poisson(ld,n)
```

```python
means['poisson'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['poisson'].iterkeys():
    means['poisson'][nn] = md['poisson'][nn].mean()
    axchisq_mu_n.plot(nn, means['poisson'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k') 
```

```python
# Generating 100 different binomial distributions with same sample mean and different sample size
# np.random.binomial(n, p, size)
md['binomial']={}
myp=0.6
myn=mymean/myp
for n in mysize:
    md['binomial'][n] = np.random.binomial(myn,myp,n)

means['binomial'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['binomial'].iterkeys():
    means['binomial'][nn] = md['binomial'][nn].mean()
    axchisq_mu_n.plot(nn, means['binomial'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k') 
```

```python
# Generating 100 different Chisquare distribution with same sample mean and different sample size
# np.random.chisquare(df, size)
# df = mean
md['chisq']={} 
df = mymean
for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size=n)
    
means['chisq'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['chisq'].iterkeys():
    means['chisq'][nn] = md['chisq'][nn].mean()
    axchisq_mu_n.plot(nn, means['chisq'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')
```

```python
# Generating 100 exponential distributions with same sample mean and different sample size
# np.random.exponential(bt, size)
np.random.seed(2015)
md['negative_binomial']={} 
myp=0.5
myn=mymean*(1-myp)/myp
for n in mysize:
    md['negative_binomial'][n] = np.random.negative_binomial(myn, myp, size=n)
    
means['negative_binomial'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['negative_binomial'].iterkeys():
    means['negative_binomial'][nn] = md['negative_binomial'][nn].mean()
    axchisq_mu_n.plot(nn, means['negative_binomial'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')
```
